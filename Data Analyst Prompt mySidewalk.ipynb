```python
import requests 
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
csv_url = "https://data.sfgov.org/api/views/nuek-vuh3/rows.csv?accessType=DOWNLOAD"
req = requests.get(csv_url)
if req.status_code == 200:
    print('Success!')
elif req.status_code == 404:
    print('Not Found.')
```

```python
url_content = req.content
csv_file = open('mySidewalk.csv', 'wb')

csv_file.write(url_content)
csv_file.close()
```

```python
df = pd.read_csv('Fire_Department_Calls_for_Service.csv',low_memory=False)
```

```python
df['Month'] = df['Response DtTm'].str.split('/', 1, expand=True)
```

```python
df.head()
```

```python
#Convert Received DtTm from object to Date time
df['Response DtTm'] = pd.to_datetime(df['Response DtTm'])
```

```python
df12 = df[(df['Response DtTm'] > '2021-5-12') & (df['Response DtTm'] <= '2022-5-12')]
```

```python
#Confirm that the correct dates were queried 
df12.sort_values(['Response DtTm'])
```

```python
#Isolate the columns needed for the analysis
ResponseBat = df12[["Battalion","Month","Response DtTm"]]
ResponseBat
```

```python
#Calculate the 90th percentile for each battalion per month
Response90 = ResponseBat.groupby(["Battalion","Month"]).quantile(0.9)
```

```python
Response90
```

```python
#Export to CSV
Response90.to_csv('90percentile.csv')
```
