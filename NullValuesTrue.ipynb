```python
arcpy.env.workspace = r"C:\Users\liammartin\Documents\ArcGIS\Projects\PythonPractice\PythonPractice.gdb"                                                                              
                                                                                                                                                                     
GDB_dir = r"C:\Users\liammartin\Documents\ArcGIS\Projects\PythonPractice\PythonPractice.gdb"                                                                               
fc_name = "medianpoints"                                                                                                                          
                                                                                                                                                                     
fc = os.path.join(GDB_dir, fc_name)
print(fc)
```

```python
sdf = pd.DataFrame.spatial.from_featureclass(fc)
sdf.head(10)
```

```python
fields = ['measure','value']
```

```python
for x in fields:
    sdf.loc[sdf[x].isnull(),x] = sdf[x].median()
sdf
```

```python
sdf.columns = map(str.upper, sdf.columns)
sdf
```

```python
out_fc = os.path.join(GDB_dir, fc_name + '__median')                                                                                                                    
sdf.spatial.to_featureclass(out_fc)
```
