```python
fcs = arcpy.ListFeatureClasses()
```

```python
FCSpath = []
GDB_dir = r"C:\Users\liammartin\Documents\Help Desk\Client Questions\Davenport\Python_workflow_data\Python_workflow_data\tabletoexcel.gdb"                                                                               
```

```python
for layer in fcs:
    x = GDB_dir + '\\'+layer
    FCSpath.append(x)
```

```python
namelist = []
```

```python
writer = pd.ExcelWriter('C:\\Users\\liammartin\\Documents\\ArcGIS\\Projects\\PythonPractice\\FCtoTable.xlsx')
```

```python
for path in FCSpath:
    name = path.split('\\')[-1]
    namelist.append(name)
    x = pd.DataFrame.spatial.from_featureclass(path)
    for sheet in namelist:
            x.to_excel(writer,sheet_name = '{}'.format(sheet))
    writer.save()
```
