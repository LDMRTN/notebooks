```python
# Copy the file path to the geodatabase that contains the R6_BigFoot_Sightings layer and paste inbetween the " " symbols below.
arcpy.env.workspace = r"C:\Users\liammartin\Documents\ArcGIS\Projects\PythonPractice\PythonPractice.gdb"

bigfoot = r"R6_BigFoot_Sightings"

arcpy.management.AddField(bigfoot, "case","SHORT")

fields = ["Year","case"]

with arcpy.da.UpdateCursor(bigfoot,fields) as cursor:
    for row in cursor:
        if row[0] >= 1861 and row[0] <= 1886:
            row[1] = 1
        elif row[0] >= 1887 and row[0] <= 1911:
            row[1] = 2
        elif row[0] >= 1912 and row[0] <= 1936:
            row[1] = 3
        elif row[0] >= 1937 and row[0] <= 1961:
            row[1] = 4
        elif row[0] >= 1962 and row[0] <= 1986:
            row[1] = 5
        elif row[0] >= 1987 and row[0] <= 2011:
            row[1] = 6
        cursor.updateRow(row)
    print("The script has finished running")
```
